## Thread avg and tail latency for compose post without istio 
This was done as a test to see how latency varies with request per second

```python
avg_lat=[1.9755,2.55,8.5890,2.54,16.26,22.77]
tail_lat=[6.0467,22.66,6.75,17.07,43.65,50.76]
x=np.arange(1,7)
plt.xticks(x,['20','50','100','200','500','1000'])
width=0.40
plt.bar(x-0.2,avg_lat,label="Average Latency",width=width)
plt.bar(x+0.2,tail_lat,label="Tail Latency",width=width)
plt.legend(['Average Latency','Tail Latency'])
```

## read timeline


```python
without_istio=[28.206,29.631,46.220,33.869,28.503]
with_istio=[52.885,53.943,68.422,67.697,81.469]
x=np.arange(1,6)
plt.xticks(x,['20','50','100','200','500'])
width = 0.40
plt.bar(x-0.2,without_istio,width=width,label="Without Istio")
plt.bar(x+0.2,with_istio,width=width,label="With Istio")
plt.legend(['Without Istio','With Istio'])
```

```python
improvement=np.subtract(with_istio,without_istio)
x=np.arange(1,6)
plt.xticks(x,['20','50','100','200','500'])
plt.plot(improvement)
```

```python
avg_improvement= np.sum(improvement)/5
print(avg_improvement)
```
