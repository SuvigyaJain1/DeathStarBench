```python
import json
import os
import subprocess
```

```python
def get_runs(current_dir):
    runs = []
    for file in os.listdir(current_dir):
        with open(os.path.join(current_dir, file), 'r') as f:
            json_object = json.load(f)
            runs.append(json_object)
    return runs
```

```python
w_istio_wo_workload = get_runs("parsedOutputs/WITH_ISTIO_WITHOUT_WORKLOAD")
```

```python
def pid_to_procname_map(runs, tracefolder):
    pid_map = {}
    count = 0
    for run in runs:
        for pid in run.keys():
            procname = subprocess.getoutput(f"python3 pid_proc.py {tracefolder}/{count}.txt {pid}")
            count += 1
            if procname in pid_map.keys():
                pid_map[procname].append(pid)
            else:
                pid_map[procname] = [pid]
        return pid_map
    return pid_map
```

```python
def pid_check(runs):
    base = {}
    base = set(runs[0].keys())
    for run in runs:
        current_set = set(run.keys())
        diff_set = current_set.difference(base)
        print(diff_set)
```
