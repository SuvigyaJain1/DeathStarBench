```python
import json
import os
```

```python
def get_runs(current_dir):
    runs = []
    for file in os.listdir(current_dir):
        with open(os.path.join(current_dir, file), 'r') as f:
            json_object = json.load(f)
            runs.append(json_object)
    return runs
```

```python
w_istio_wo_workload = get_runs("WITH_ISTIO_WITHOUT_WORKLOAD")
```

```python
def pid_check(runs):
    base = {}
    base = set(runs[0].keys())
    for run in runs:
        current_set = set(run.keys())
        diff_set = current_set.difference(base)
        print(diff_set)
```

```python
pid_check(w_istio_wo_workload)
```
